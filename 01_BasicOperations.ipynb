```python
import tensorflow as tf
#The notion of tensorflow is first constructing a graph (numbers & operations),
#then later we feed the data (using feed_dict) when running the constructed graph (using sess.run).
a = tf.constant(2)
b = tf.constant(3)
add = tf.add(a, b)
with tf.Session() as sess:
    print("a: {}".format(a))
    print("a: {}".format(sess.run(a)))
    print("a: {}".format(a.eval()))
    print("a+b: {}".format(sess.run(a+b)))
    print("a+b: {}".format(sess.run(add)))
```

```python
a = tf.placeholder(tf.int16)
b = tf.placeholder(tf.int16)
add = tf.add(a, b)
mul = tf.multiply(a, b)
with tf.Session() as sess:
    print("a: {}".format(a))
    print("Addition with variables: {}".format(sess.run(add, feed_dict={a:3, b:2})))
    print("Multiply with variables: {}".format(sess.run(mul, feed_dict={a:3, b:2})))
```

```python
A = tf.constant([[3., 3.]])
B = tf.constant([[2.],[2.]])
product = tf.matmul(A, B)
with tf.Session() as sess:
    print("A: {}".format(A))
    print("A: {}".format(sess.run(A)))
    result = sess.run(product)
    print(result)
```
